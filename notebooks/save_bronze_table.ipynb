```python
from pyspark.sql.functions import col,explode
 
```

```python
df = spark.read.json("abfss://job_market@onelake.dfs.fabric.microsoft.com/lakehouse.Lakehouse/Files/data_f3c6bafd-25cb-4df3-a822-33dc976e40e8_43bb0728-a2d3-4da2-b05c-504e61432ba3.json",multiLine=True)
```

```python
display(df.limit(2))
```

```python
exploded_df = df.select(explode(col("jobs")).alias("job"))
```

```python
display(exploded_df)
```

```python
job_falt_df = exploded_df.select(
    col("job.company_name").alias("company_name"),
    col("job.title").alias("job_title"),
    col("job.category").alias("category"),
    col("job.job_type").alias("job_type"),
    col("job.publication_date").alias("job_date"),
    col("job.candidate_required_location").alias("location"),
    col("job.salary").alias("salary"),
    




)
```

```python
display(job_falt_df.limit(2))
```

```python

job_falt_df.write.format("delta").mode("append").saveAsTable("bronze_jobs")
```
